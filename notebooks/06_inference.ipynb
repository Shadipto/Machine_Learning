# Phase 6 - Inference Pipeline

Load the saved vectorizer and best model, then define a simple `predict_text` helper for raw input text.

```python
import pickle
import re
import string
from pathlib import Path

import pandas as pd

print("Inference imports loaded")
```

## Loading Saved Vectorizer and Best Model for Inference

```python
models_dir = Path("../models")
vectorizer_path = models_dir / "tfidf_vectorizer.pkl"
model_path = models_dir / "voting_classifier.pkl"

with open(vectorizer_path, "rb") as file_handle:
    vectorizer = pickle.load(file_handle)

with open(model_path, "rb") as file_handle:
    model = pickle.load(file_handle)

print("Loaded vectorizer:", vectorizer_path)
print("Loaded model:", model_path)
```

## Preparing Functions for Model Inference



```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", " ", text)
    text = re.sub(r"<.*?>", " ", text)
    text = text.translate(str.maketrans("", "", string.punctuation))
    text = re.sub(r"\d+", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text


def predict_text(text):
    cleaned_text = clean_text(text)
    features = vectorizer.transform([cleaned_text])
    prediction = model.predict(features)[0]
    return prediction


print("predict_text is ready")
```

## Running Sample Predictions Using the Trained Model

```python
sample_texts = [
    "I feel like nothing is worth it anymore and I don't want to keep going.",
    "I had a good day with my friends and feel thankful.",
    "Sometimes I just want to disappear and not deal with anything.",
]

for idx, sample in enumerate(sample_texts, start=1):
    print(f"Sample {idx}: {predict_text(sample)}")
    print(f"Text: {sample}")
    print("-" * 60)
```
